```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/archive (8).zip")
```

```python
print(df.info())
```

```python
print(df.describe(include='all'))
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
print("\n Histograms")
sns.set(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(10, 8))
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.histplot(df[col].dropna(), kde=True, ax=ax, bins=30)
    ax.set_title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
print("\n Boxplots")
fig, axes = plt.subplots(2, 3, figsize=(11, 6))
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
print("\n Correlation Heatmap")
corr = df[numeric_cols + ['Survived']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
print("\n📌 Pairplot")
sns.pairplot(df[numeric_cols + ['Survived']], hue='Survived', palette='Set1')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival Count by Gender")
plt.show()
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.show()
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title("Fare Distribution by Survival")
plt.show()
sns.boxplot(data=df, x='Pclass', y='Age')
plt.title("Age Distribution Across Passenger Classes")
plt.show()
sns.heatmap(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
